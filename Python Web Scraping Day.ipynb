```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.imdb.com/search/title/?title_type=tv_series&countries=tr&languages=tr&count=250&start="
```

```python
all_links=[]
for i in range(0,2250,250):
    all_links.append(url+str(i))
all_links
```

```python
url_inside=[]

for i in all_links:
    page = requests.get(i)
    pages = BeautifulSoup(page.text, 'html.parser')
    results = pages.find(class_="lister-list").find_all("a")
    for result in results:
        if result["href"].__contains__("title") and not(result["href"].__contains__("vote")) and not (result["href"].__contains__("plotsummary")):
            url_inside.append("https://imdb.com"+result["href"])

        
    
```

```python
url_inside = list(dict.fromkeys(url_inside))
url_inside = list(set(url_inside))
len(url_inside)
```
